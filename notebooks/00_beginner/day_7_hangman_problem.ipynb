## Hangman Game Designing using Python

```python
import random

stages = [
'''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', 
'''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', 
'''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', 
'''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', 
'''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', 
'''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', 
'''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

word_list = ["world","butterfly","everything"]

chosen_word = word_list[random.randint(0,2)]

print(chosen_word)

guess_list = ["_" for i in range(len(chosen_word))] 

lives = 6

while ("_" in guess_list):
    
    guess = input("Guess a letter").lower()

    # rpls_guess_list = [pos for pos, char in enumerate(chosen_word) if char == guess]

    #for i in rpls_guess_list:
        # guess_list[i] = guess

    for position in range(len(chosen_word)):
        letter = chosen_word[position]
        if letter == guess:
            guess_list[position] = letter
   
    if guess not in chosen_word:
        #print(stages[lives])
        lives -= 1
        
    
    print(guess_list) 
    print("lives left: ", lives)

    if lives == 0:
        print("You lose!")
        break
        
```
